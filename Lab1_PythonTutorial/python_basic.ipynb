## List Tuple Set Dict
1. 列表（List）
特点：有序、可变、允许重复元素。
语法：用方括号 [] 定义，元素用逗号分隔。


适用场景：存储需要保持顺序且可能需要修改的数据。
2. 字典（Dict）
特点：无序（Python 3.7+ 后保持插入顺序）、键值对（key-value）结构、键唯一且不可变（如字符串、数字、元组）。
语法：用花括号 {} 定义，键值对用冒号分隔，键值对之间用逗号分隔。


适用场景：快速查找和存储关联数据（如配置、统计）。
3. 元组（Tuple）
特点：有序、不可变（元素不能修改）、允许重复元素。
语法：用圆括号 () 定义（可省略括号），元素用逗号分隔。


适用场景：存储不可变数据（如坐标、常量）或作为字典的键。
4. 集合（Set）
特点：无序、元素唯一且不可变（无重复值）、支持数学集合操作（如并集、交集）。
语法：用花括号 {} 或 set() 函数定义（空集合必须用 set()）。

适用场景：去重、成员检测（快速判断元素是否存在）、集合运算。
核心区别总结

选择建议
需要顺序和可修改：用 list。
需要键值对和快速查找：用 dict。
需要不可变数据：用 tuple。
需要去重或集合操作：用 set。

```python
# -*- coding: UTF-8 -*-

list = [ 'hello', 2 ,'you', 'hh', 666 ]
tinylist = [123, 'go']

## list 
print (list)              # 输出完整列表
print (list[0])            # 输出列表的第一个元素
print (list[1:3])          # 输出第二个至第三个元素 
print (list[2:] )          # 输出从第三个开始至列表末尾的所有元素
print (tinylist * 2 )      # 输出列表两次
print (list + tinylist)    # 打印组合的列表


## tuple
tuple = ( 'hello', 2 ,'you', 'hh', 666 )
tinytuple = (123, 'go')
print (tuple)              # 输出完整列表
print (tuple[0])            # 输出列表的第一个元素
print (tuple[1:3])          # 输出第二个至第三个元素 
print (tuple[2:] )          # 输出从第三个开始至列表末尾的所有元素
print (tinytuple * 2 )      # 输出列表两次
print (tinytuple + tuple)    # 打印组合的列表




## dict  
dict = {}
dict['one'] = "This is one"
dict[2] = "This is two"
 
tinydict = {'name': 'anna ','age':20, 'height': '180'}
print (dict['one'])          # 输出键为'one' 的值
print (dict[2])              # 输出键为 2 的值
print (tinydict)           # 输出完整的字典
print (tinydict.keys())      # 输出所有键
print (tinydict.values())   # 输出所有值
```

class 
基本概念
类（Class）：是对象的抽象定义，描述了对象的属性和行为。 
对象（Object）：是类的实例，通过类创建。
属性（Attribute）：类中的变量，存储对象的数据。
方法（Method）：类中的函数，定义对象的行为。

为什么使用类？
封装：将数据和操作捆绑，保护数据不被外部随意修改。
继承：实现代码复用和扩展，建立类的层次结构。
多态：不同子类可以对同一方法有不同实现。

```python
class Person:
    # 类属性（所有实例共享）
    species = "Homo sapiens"

    def __init__(self, name, age):
        # 实例属性（每个实例独立）
        self.name = name
        self.age = age

    # 实例方法
    def greet(self):
        return f"Hello, my name is {self.name} and I'm {self.age} years old."

# 实例化
p1 = Person("Alice", 30)  # 创建 Person 类的实例
p2 = Person("Bob", 25)    # 创建另一个实例

print(p1.name)    # 输出: Alice
print(p1.greet()) # 输出: Hello, my name is Alice and I'm 30 years old.
```

```python
class Student(Person):
    def __init__(self, name, age, school):
        super().__init__(name, age)  # 调用父类的构造函数
        self.school = school

    def study(self):
        return f"{self.name} is studying at {self.school}."
s1 = Student ("Charlie", 20, "MIT")
print (s1.greet ()) # 继承自 Person 类
print (s1.study ()) # 子类特有的方法
```
